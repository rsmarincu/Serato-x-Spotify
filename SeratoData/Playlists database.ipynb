```python
from bs4 import BeautifulSoup as bs
import requests
from pprint import pprint
import re
import json
from pprint import pprint
```

```python
SERATO_BASE_URL = "https://serato.com"
SERATO_GENRES_URL = "https://serato.com/playlists/browse/genres"
```

```python
genre_page = requests.get(SERATO_GENRES_URL)
soup = bs(genre_page.content)

genres = []

for link in soup.findAll("a", class_="genre-link",href=True):
    genres.append((link.span.text,link['href']))

print(genres)
```

```python
playlist_list = []

for genre in genres:
    hrefs = []
    QUERY_URL = SERATO_BASE_URL+genre[1]
    playlists_page = requests.get(QUERY_URL)
    soup = bs(playlists_page.content)
    for playlist in soup.findAll("a", class_="playlist-title",href=True):
        playlist_name = re.split(" \\| ",playlist.text)
        hrefs.append(playlist['href'])
    playlist_list.append(hrefs)

```

```python
pprint (playlist_list)
```

```python
def createPlaylist(playlist_url,name,genre,dj,createJson=False):
    tracks = []
    QUERY_URL = SERATO_BASE_URL+playlist_url
    tracklist_page = requests.get(QUERY_URL)
    soup = bs(tracklist_page.content)
    for tracklist in soup.findAll("div", class_="playlist-trackname"):
        full_track = tracklist.text
        if full_track.isspace():
            break
        full_track = re.split("\n                    ",full_track)
        full_track = re.split("\t\t\t\t",full_track[1])
        full_track = re.split(" - ",full_track[0])
        artist = full_track[0]
        try:
            trackname = full_track[1]
        except IndexError:
            trackname = artist
        clean_name = {
            "artist" : artist,
            "name" : trackname
        }
        tracks.append(clean_name)
    if len(tracks) == 0:
        return None
    playlist = {
        "Name" : name,
        "Genre" : genre,
        "DJ" : dj,
        "Length" : len(tracks),
        "Tracks" : tracks
    }
    if createJson:
        with open ("{}.json".format(name),'w') as outfile:
            json.dump(playlist, outfile)
    
    return playlist
```

```python
pprint(createPlaylist(playlist_list[0][0],"mobjey","RB","qsd"))
```

```python
dataframe = []

for i,genre in enumerate (genres):
    for playlist in playlist_list[i]:
        unformated = re.split("/",playlist)
        dj = unformated[2]
        name = unformated[3]
        a = createPlaylist(playlist,name,genre[0],dj)
        if a:
            dataframe.append(a)
    if i==1:
        break
```
