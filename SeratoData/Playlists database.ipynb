```python
from bs4 import BeautifulSoup as bs
import requests
from pprint import pprint
import re
import json
from pprint import pprint
```

```python
SERATO_BASE_URL = "https://serato.com"
SERATO_GENRES_URL = "https://serato.com/playlists/browse/genres"
```

```python
genre_page = requests.get(SERATO_GENRES_URL)
soup = bs(genre_page.content)

genres = []

for link in soup.findAll("a", class_="genre-link",href=True):
    genres.append((link.span.text.strip(),link['href']))

print(genres)
```

```python
pprint(genres)
```

```python
playlist_list = []

for genre in genres:
    hrefs = []
    QUERY_URL = SERATO_BASE_URL+genre[1]
    playlists_page = requests.get(QUERY_URL)
    soup = bs(playlists_page.content)
    for playlist in soup.findAll("a", class_="playlist-title",href=True):
        playlist_name = re.split(" \\| ",playlist.text)
        hrefs.append(playlist['href'])
    playlist_list.append(hrefs)

```

```python
pprint (playlist_list)
```

```python
def createPlaylist(playlist_url,name,genre,dj,createJson=False):
    tracks = []
    QUERY_URL = SERATO_BASE_URL+playlist_url
    tracklist_page = requests.get(QUERY_URL)
    soup = bs(tracklist_page.content)
    for tracklist in soup.findAll("div", class_="playlist-trackname"):
        full_track = tracklist.text
        if full_track.isspace():
            break
        full_track = re.split("\n                    ",full_track)
        full_track = re.split("\t\t\t\t",full_track[1])
        full_track = re.split(" - ",full_track[0])
        artist = full_track[0]
        try:
            trackname = full_track[1]
        except IndexError:
            trackname = artist
        trackname = trackname.replace('(Dirty)','')
        trackname = trackname.replace('(Clean)','')
        trackname = trackname.strip()
        clean_name = {
            "artist" : artist,
            "name" : trackname
        }
        tracks.append(clean_name)
    if len(tracks) == 0:
        return None
    playlist = {
        "Name" : name,
        "Genre" : genre,
        "DJ" : dj,
        "Length" : len(tracks),
        "Tracks" : tracks
    }
    if createJson:
        with open ("{}.json".format(name),'w') as outfile:
            json.dump(playlist, outfile)
    
    return playlist
```

```python
SPOTIFY_AUDIO_FEATURES='https://api.spotify.com/v1/audio-features/'

HEADERS_SEARCH = {
    "Authorization": "Bearer {}".format(access_token)
}

def getPlaylistFeatures(playlist):
    jsonplaylist = playlist.to_json(orient='index')
    jsonplaylist = json.loads(jsonplaylist)
    featurelist = []
    for track in jsonplaylist['Tracks']:        
        DATA_SEARCH = {
            "q":track['name']+' '+track['artist'],
            'type':'track,artist'    
        }
        song = requests.get(SPOTIFY_SEARCH_API, params=DATA_SEARCH,headers=HEADERS_SEARCH)
        song = song.json()
        try:
            global trackid 
            trackid = song['tracks']['items'][0]['id']
        except IndexError:
            pass
        url = SPOTIFY_AUDIO_FEATURES+trackid
        features = requests.get(url, headers=HEADERS_SEARCH)
        features = features.json()
        featurelist.append(features)
    return featurelist

f = getPlaylistFeatures(testdf)
```
